```python
import yaml
from pyspark.sql.functions import col, avg, round
```

```python
with open("config.yaml", "r") as file:
    config = yaml.safe_load(file)
```

```python
silver = spark.read.parquet(
    config["silver_storage_base_path"]
)
```

```python
display(
    silver.groupBy("city", "year").agg(
        round(
            avg("temperature_mean"),
            2
        ).alias("average_temperature")
    ).orderBy("city", "year")
)
```
