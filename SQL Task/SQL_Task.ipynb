## 1.  Table name: Programmer

```sql
CREATE OR REPLACE TABLE programmer(
  pname VARCHAR(25) NOT NULL,
  dob DATE NOT NULL,
  doj DATE NOT NULL,
  gender CHAR(1) NOT NULL,
  prof1 VARCHAR(10) NOT NULL,
  prof2 VARCHAR(10) NOT NULL,
  salary INT NOT NULL,
  PRIMARY KEY (pname)
);
```

```sql
CREATE OR REPLACE TABLE studies(
  pname VARCHAR(25) NOT NULL,
  institute VARCHAR(20) NOT NULL,
  course VARCHAR(20) NOT NULL,
  course_fee INT NOT NULL,
  FOREIGN KEY (pname) REFERENCES programmer(pname)
);
```

## 3. Table name: Software

```sql
CREATE OR REPLACE TABLE software(
  pname VARCHAR(25) NOT NULL,
  title VARCHAR(15) NOT NULL,
  develop_in VARCHAR(8) NOT NULL,
  scost FLOAT NOT NULL,
  dcost FLOAT NOT NULL,
  sold INT NOT NULL,
  FOREIGN KEY (pname) REFERENCES programmer(pname)
);
```

## 21. Display the details of the software developed in DBASE by Male Programmers, who belong to the institute in which most number of Programmers studied.  

1. Sub Query

```sql
select * from software 
where develop_in = "DBASE" 
and pname in (
  select pname from programmer 
  where gender = "M" and pname in (
    select pname from studies 
    where institute in (
      select institute from studies 
      group by institute 
      order by count(institute) 
      desc limit 1
    )
  )
);
```

2. Using Joins

```sql
SELECT S.PNAME, P.GENDER, ST.INSTITUTE, S.TITLE, S.DEVELOP_IN, S.SCOST, S.DCOST, S.SOLD FROM SOFTWARE S 
INNER JOIN PROGRAMMER P 
INNER JOIN STUDIES ST ON (
  S.PNAME = P.PNAME AND P.PNAME = ST.PNAME
  ) WHERE S.PNAME IN (
    SELECT PNAME FROM STUDIES 
    WHERE INSTITUTE IN (
      SELECT INSTITUTE FROM STUDIES 
      GROUP BY INSTITUTE 
      ORDER BY COUNT(INSTITUTE) DESC LIMIT 1
    )
  )
  AND S.DEVELOP_IN = 'DBASE' AND P.GENDER = 'M';
```

## 35. Create a T-SQL stored procedure named CalculateStoreRevenue that takes two parameters: 

- **@StartDate** - Date (to specify the start date for the date range), 
- **@EndDate** - Date (to specify the end date for the date range). 
- The stored procedure should calculate and return the total revenue generated by each store within the specified date range. 
- The result set should include the StoreName, Location, and TotalRevenue. 

**Stored Procedure Requirements:** The stored procedure should handle cases where there are no orders for a store within the specified date range (in such cases, the TotalRevenue should be 0). Use proper error handling to handle any potential errors or issues. Ensure that the stored procedure is optimized for performance. 

Online compiler link: https://dbfiddle.uk/kYNo1I4-

```sql
%skip
CREATE TABLE stores(
  store_id INT PRIMARY KEY,
  store_name VARCHAR(20),
  location VARCHAR(15)
);

CREATE TABLE orders(
  order_id INT PRIMARY KEY,
  store_id INT,
  order_date DATE,
  total_amount DECIMAL(10, 2),
  FOREIGN KEY (store_id) REFERENCES stores(store_id)
);

INSERT INTO stores VALUES
(1, 'TechMart', 'Chennai'),
(2, 'GadgetHub', 'Bangalore'),
(3, 'ElectroWorld', 'Hyderabad'),
(4, 'SmartStore', 'Mumbai'),
(5, 'DigitalZone', 'Delhi');

INSERT INTO orders VALUES
(101, 1, '2024-01-05', 15000.00),
(102, 1, '2024-01-10', 20000.00),
(103, 2, '2024-01-07', 12000.00),
(104, 2, '2024-02-15', 18000.00),
(105, 3, '2024-01-20', 22000.00),
(106, 3, '2024-03-01', 25000.00),
(107, 4, '2024-01-25', 30000.00),
(108, 4, '2024-02-05', 27000.00);
```

```sql
%skip
CREATE PROCEDURE calculate_store_revenue
  @StartDate DATE,
  @EndDate DATE
AS
BEGIN
  BEGIN TRY
    SELECT 
      s.store_name, 
      s.location, 
      SUM(o.total_amount) as total_revenue
    FROM stores s
    LEFT JOIN orders o
      ON s.store_id = o.store_id
    WHERE o.order_date BETWEEN @StartDate AND @EndDate
    GROUP BY s.store_name, s.location;
  END TRY

  BEGIN CATCH
    SELECT ERROR_MESSAGE() AS ErrorMessage;
  END CATCH
END;
```

## 40. Merit is booking Rooms for Employees in a Hotel, who are interested in North India Holiday Vacation Trip. How can we create and utilize a User Defined Function in SQL Server to calculate the total cost of all hotel room reservations for a specific room type, taking into account the ticket type as a parameter input?. Retrieve the total cost of all reservations for ticket type 'Standard'. 

Online compiler URL: https://dbfiddle.uk/HSdcP_If

### Creating necessary tables

```sql
%skip
-- 1. Rooms table
CREATE TABLE rooms (
    room_id INT PRIMARY KEY,
    room_type VARCHAR(10),
    base_price FLOAT
);

-- 2. Tickets table
CREATE TABLE tickets (
    ticket_type VARCHAR(8) PRIMARY KEY,
    ticket_price FLOAT
);

-- 3. Reservations table
CREATE TABLE reservations (
    res_id INT PRIMARY KEY,
    employee_id INT,
    room_type VARCHAR(10),
    ticket_type VARCHAR(8),
    stay_duration INT,
    FOREIGN KEY (ticket_type) REFERENCES tickets(ticket_type)
);

-- Inserting into tables

-- 1. Rooms table
INSERT INTO rooms (room_id, room_type, base_price) VALUES
(101, 'Deluxe', 4000.00),
(102, 'Suite', 6000.00),
(103, 'Deluxe', 4200.00),
(104, 'Suite', 6500.00),
(105, 'Deluxe', 4500.00),
(106, 'Suite', 7000.00);

-- 2. Tickets table
INSERT INTO tickets (ticket_type, ticket_price) VALUES
('STANDARD', 500.00),
('PREMIUM', 1000.00);

-- 3. Reservations table
INSERT INTO reservations (res_id, employee_id, room_type, ticket_type, stay_duration) VALUES
(1, 201, 'Deluxe', 'STANDARD', 2),
(2, 202, 'Suite', 'PREMIUM', 3),
(3, 203, 'Suite', 'STANDARD', 1),
(4, 204, 'Deluxe', 'PREMIUM', 4),
(5, 205, 'Deluxe', 'STANDARD', 2),
(6, 206, 'Suite', 'STANDARD', 5),
(7, 207, 'Suite', 'PREMIUM', 2),
(8, 208, 'Deluxe', 'PREMIUM', 3);

-- Creating a user-defined function to retrieve total cost
CREATE FUNCTION calculate_total_cost (
    @ticket_type VARCHAR(8)
)
RETURNS DECIMAL(10,2)
AS
BEGIN
    DECLARE @result DECIMAL(10,2);

    SELECT 
        @result = SUM(
            (rooms.base_price * reservations.stay_duration) 
            + tickets.ticket_price
        )
    FROM reservations
    JOIN rooms 
        ON reservations.room_type = rooms.room_type
    JOIN tickets 
        ON reservations.ticket_type = tickets.ticket_type
    WHERE tickets.ticket_type = @ticket_type;

    RETURN ISNULL(@result, 0);
END;

-- Calling the function
SELECT dbo.calculate_total_cost('STANDARD') AS TotalStandardCost;
```
